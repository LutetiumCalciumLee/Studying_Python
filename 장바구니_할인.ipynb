- 물건 가격이 10,000원 이상이면 10% 할인이 적용됩니다.​

- 물건 가격이 10,000원 미만이면 할인이 적용되지 않습니다.​

- 여러 개의 물건을 샀을 때 최종 지불 금액을 계산하세요.

```python
import random
```

```python
ShoppingPrice = []
```

```python
product1price = random.randrange(1000,20000,100)
product2price = random.randrange(1000,20000,100)
product3price = random.randrange(1000,20000,100)
product4price = random.randrange(1000,20000,100)
product5price = random.randrange(1000,20000,100)
```

```python
ShoppingPrice.append(product1price)
ShoppingPrice.append(product2price)
ShoppingPrice.append(product3price)
ShoppingPrice.append(product4price)
ShoppingPrice.append(product5price)
```

```python
print(ShoppingPrice)
```

```python
total = 0
```

```python
for price in ShoppingPrice:
  if price >= 10000 :
    price *= 0.9
  total += price
print (total)
```
