<요구사항>​

1. 사용자로부터 비밀번호를 입력 받습니다.​
2. 다음 조건을 검사하여 각각 점수를 부여합니다:​

 ·비밀번호 길이가 8자 이상이면 2점​

 ·숫자를 하나 이상 포함하면 2점​

 ·대문자를 포함하면 1점​

 ·소문자를 포함하면 1점​

3. 각 조건은 별도의 함수로 검사하며, 점수와 피드백 메시지를 반환하도록 만듭니다.​

4. 모든 조건을 검사한 뒤, 총점을 계산하고 다음 기준으로 강도를 출력합니다:​

 ·총점이 6점 이상: 비밀번호 강도: 강함​

 ·총점이 4~5점: 비밀번호 강도: 보통​

 ·총점이 3점 이하: 비밀번호 강도: 약함​

5. 각 조건별 평가 메시지를 함께 출력합니다.​
(예: "✔ 숫자 포함", "✖ 대문자 없음" 등)​

```python
PW = input("""비밀번호를 생성하시오
 - 8자 이상
 - 숫자를 최소 하나 포함
 - 대문자 포함
 - 소문자 포함 \n""")
```

```python
point = []
```

```python
if len(PW) >= 8:
    point.append(2)
    print("✔ 8자 이상")
else:
    print("✖ 8자 이상 아님")
```

```python
if any(char.isdigit() for char in PW):
    point.append(2)
    print("✔ 숫자 포함")
else:
    print("✖ 숫자 없음")
```

```python
if any(char.isupper() for char in PW):
    point.append(1)
    print("✔ 대문자 포함")
else:
    print("✖ 대문자 없음")
```

```python
if any(char.islower() for char in PW):
    point.append(1)
    print("✔ 소문자 포함")
else:
    print("✖ 소문자 없음")
```

```python
print(point)
```

```python
TotalPoint = 0
for i in point:
  TotalPoint += i

if TotalPoint >= 6 :
  print ("비밀번호 강도 : 강함")

elif TotalPoint >= 4 and TotalPoint <= 5 :
  print ("비밀번호 강도 : 보통")

elif TotalPoint <= 3 :
  print ("비밀번호 강도 : 약함")
```
